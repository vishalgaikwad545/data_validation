```python
import pandas as pd
import random

# Function to generate random data
def generate_data(num_rows=100):
    data = {
        "drug_id": [random.randint(1000000, 9999999) for _ in range(num_rows)],
        "practitioner_id": [random.randint(10000000, 99999999) for _ in range(num_rows)],
        "plan_id": [random.randint(100000, 999999) for _ in range(num_rows)],
        "zip_code": [random.randint(100, 999) for _ in range(num_rows)],  # 3-digit ZIP
    }
    return pd.DataFrame(data)

# Generate DataFrame
df = generate_data(1000)  # change number of rows if needed

# Save to Parquet
df.to_parquet("sample_data.parquet", index=False)

print("✅ Parquet file 'sample_data.parquet' created successfully.")
```
